````python
# Solution

def rgb(r, g, b):
    rgb = [r, g, b]
    for i, n in enumerate(rgb):
        if 0 > n:
            rgb[i] = 0
        elif 255 < n:
            rgb[i] = 255
    return '{:02X}{:02X}{:02X}'.format(*rgb)
````
